## Sigma0 Analysis - PNB

Pré-processamento: Orbit File -> Calibration -> Speckle Filtering -> Terrain Correction -> Spatial Subset

```python
# Importações
# Bibliotecas básicas
import rasterio as rst
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd

# Rasterio
import rasterio as rst
from rasterio.plot import show
# Raster stats
from rasterstats import zonal_stats
from scipy.stats import iqr
# Processamento de dados
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
# Áreas de estudo no Parque Nacional de Brasília
PNB = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/PNB_32723.GEOJSON')
form_florestal = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/formacao_florestal.GEOJSON')
form_savanica = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/formacao_savanica.GEOJSON')
form_campestre = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/formacao_campestre.GEOJSON')
agua = gpd.read_file('C:/Users/jales/Desktop/PNB/classes/agua.GEOJSON')

# Importação da imagem preprocessada
with rst.open('E:/SBSR_data/TIFF/S1A_IW_GRDH_1SDV_20190911_Orb_Cal_Spk_TC.tif', 'r') as image:
    VH = image.read(1)
    VV = image.read(2)
```

```python
# Exportação da imagem final
out_meta.update({"driver": "GTiff",
                 "height": VH.shape[0],
                 "width": VH.shape[1],
                 "transform": image.transform})
    
with rst.open('E:/SBSR_data/TIFF/S1A_IW_GRDH_1SDV_20190911_Orb_Cal_Spk_TC_VH.tif', "w", **out_meta) as dest:
    dest.write(VH, 1)
```

```python
# Exportação da imagem final
out_meta.update({"driver": "GTiff",
                 "height": VV.shape[0],
                 "width": VV.shape[1],
                 "transform": image.transform})
    
with rst.open('E:/SBSR_data/TIFF/S1A_IW_GRDH_1SDV_20190911_Orb_Cal_Spk_TC_VV.tif', "w", **out_meta) as dest:
    dest.write(VV, 1)
```

```python
VH_flat = VH.flatten()
VV_flat = VV.flatten()
img_data = pd.DataFrame({'VH': VH_flat,'VV': VV_flat})
```

```python
fig = plt.figure()
ax = fig.add_subplot()

x_VV = img_data['VV']
y_VH = img_data['VH']

ax.scatter(x_VV, y_VH)

ax.set_xlabel('Sigma0_VV')
ax.set_ylabel('Sigma0_VH')

plt.show()
```

```python
print(f'VH_Max = {VH.max()}')
print(f'VH_min = {VH.min()}')
print()
print(f'VV_Max = {VV.max()}')
print(f'VV_min = {VV.min()}')
```

```python
# Canal VH
fig = plt.figure(figsize=(20,10))
plt.imshow(VH, vmin=np.percentile(VH, 2), vmax=np.percentile(VH, 98))
plt.title('Sigma0_VH')
```

```python
# Canal VV
fig = plt.figure(figsize=(20,10))
plt.imshow(VV, vmin=np.percentile(VH, 2), vmax=np.percentile(VH, 98))
plt.title('Sigma0_VV')
```
